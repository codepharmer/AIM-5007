```python
import numpy as np
import matplotlib.pyplot as plt
```

**Below are the graphs for the activation functions included in problem set 4 along with graphs for their respective first derivatives.**

## 1. Hyperbolic Tangent (tanh) 

```python
def tanh(x):
    return (np.e**x - np.e**(-x))/ (np.e**x+np.e**(-x))
def dtanh(x):
    return 1 - (np.e**x-np.e**-x)**2/(np.e**x+np.e**-x)**2
```

```python
#we'll use this to plot our activation functions
def function_plotter(f, fun_name, k = None):
    x_vals = np.linspace(-5,5,500)
    if k == None:
        y_vals = [f(x) for x in x_vals]
    else:
        y_vals = [f(k, x) for x in x_vals]
    plt.plot(x_vals, y_vals)
    plt.title(size=14, label=fun_name)
    plt.grid()
    plt.show()
```

```python
function_plotter(tanh, "tanh(x)")
function_plotter(dtanh, "tanh'(x)")
```

## 2. Linear Function

```python
def linear(k, x):
    return k*x
    
def dlinear(k, x):
    return k
```

```python
#assign to k a random constant
k = np.random.randint(1,100)
function_plotter(linear, "linear(x)", k)
function_plotter(dlinear, "linear'(%s*x)" % (k), k)
```

## 3. Step Function

```python
def step(x):
    if x <= 0:
        return 0
    if x > 0:
        return 1
    
def dstep(x):
    if x != 0:
        return 0
    else:
        return None
```

```python
function_plotter(step, "step(x)")
### Note: disconinuity exists at x=0, but a point won't be visible as it is an infinitesmal
print('')
function_plotter(dstep, "step'(x)")
```

## 4. Rectified Linear Unit (ReLU)

```python
def relu(x):
    if x <= 0:
        return 0
    if x > 0:
        return x
    
def drelu(x):
    if x <= 0:
        return 0
    else:
        return 1
```

```python
function_plotter(relu, "ReLU(x)")
### Note: disconinuity exists at x=0, but a point won't be visible as it is an infinitesmal
function_plotter(drelu, "ReLU'(x)")
```

## 5. Softplus

```python
def softplus(x):
    return np.log(1+np.e**x)
def dsoftplus(x):
    return np.e**x / (1+np.e**x)
```

```python
function_plotter(softplus, "softplus(x)")
function_plotter(dsoftplus, "softplus'(x)")
```
